```python
# This is a Python script that defines a function to concatenate a person's first name and last name.

# Function to concatenate first name and last name
def fullname(firstname, lastname):
    return firstname + " " + lastname

firstname = input("Enter your first name: ")
lastname = input("Enter your last name: ")
print(fullname(firstname, lastname))
```

```python
# Description: This program takes a string as input from the user and prints the alternative characters of the string.
def string_alternative(input_string):
    return input_string[::2]

def main():
    """
    This function takes user input, calls the string_alternative function, and prints the result.
    """
    input_string = input("Enter a string: ")
    print(string_alternative(input_string))
main()
```

```python
# Write a python program to find the wordcount in a file (input.txt) for each line and then print the output.
# o Finally store the output in output.txt file. 

# Function to count the number of words in a file
def word_count(input_file, output_file):
    word_dict = {}
    # Open the input file in read mode
    with open(input_file, 'r') as file:
        lines = file.readlines()
    # Open the output file in write mode
    with open(output_file, 'w') as file:
        for line in lines:
            file.write(line)
            words = line.split()
            # Count the number of words in the line
            for word in words:
                # If the word is already in the dictionary, increment the count
                if word in word_dict:
                    word_dict[word] += 1
                else:
                    word_dict[word] = 1
        # Write the word count to the output file
        file.write("\nWord_Count:\n")
        for word, count in word_dict.items():
            file.write(f"{word}: {count}\n")



# Call the function
word_count('input.txt', 'output.txt')

```

```python
# Write a program, which reads heights (inches.) of customers into a list and convert these
# heights to centimeters in a separate list using:
# 1) Nested Interactive loop.
# 2) List comprehensions

#`Finally, print the heights in inches and centimeters.`
def height_conversion():
    heights = []
    # Interactive loop
    while True:
        height = input("Enter the height in inches: ")
        if height == "":
            break
        heights.append(int(height))
    
    # Nested Interactive loop
    heights_cm = []
    for height in heights:
        heights_cm.append(height * 2.54)
    print(heights)
    
    # List comprehensions
    heights_cm = [round((height * 2.54), 2) for height in heights]
    print(heights_cm)

height_conversion()
```
