```python
# getting string from user
inp = input("Enter a Stirng: ")

print(inp)

# remove 2 chars from the input string
modified_string = inp[:-2]  #slicing string to remove last 2 chars
print(modified_string)

#reverse the modified string
rev_str = modified_string[::-1] # using python slicing to reverse string 
print(rev_str)
```

```python
#– Take two numbers from user and perform at least 4 arithmetic operations on them

# getting num1 and num2 from user.
num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

# printing input numbers
print(f"num1: {num1} \nnum2: {num2}")

#Arthematic OPerations 
print(f"Addition operation: {num1 + num2}")
print(f"Substraction operation: {num1 - num2}")
print(f"Multiplication operation: {num1 * num2}")
print(f"Division operation: {num1 / num2}")
```

```python
#Use the if statement conditions to write a program to print the letter grade based on an input class score. Use the grading scheme we are using in this class.
score = int(input("Enter Score: "))

if score >= 90: # check if score greater than 90 
    print("Grade is A")
elif score >= 80 and score < 90: # check if score between 
    print("Grade is B")
elif score >= 70 and score < 80:
    print("Grade is C")
else:
    print("Grade is D")
```
